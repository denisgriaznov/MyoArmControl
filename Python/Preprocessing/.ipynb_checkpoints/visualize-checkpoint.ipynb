```python
def sigmaValue(inp):
    length = inp.shape[0]
    sigma = 0
    for i in inp:
        sigma += i*i
    sigma /= length
    sigma = math.sqrt(sigma)
    return sigma
```

```python
def meanValue(inp):
    mean = 0
    length = inp.shape[0]
    for i in inp:
        mean += i
    mean /= length
    return mean
```

```python
def standartization(inp):
    mean = meanValue(inp)
    sigma = sigmaValue(inp)
    length = inp.shape[0]
    output = np.zeros(length)
    for i in range(length):
        output[i] = (inp[i] - mean)/sigma
    return output
```

```python
def movingAverageFilter(inp,n):
    inp = abs(inp)
    length = inp.shape[0]
    output = np.zeros(length)
    for i in range(length):
        current = 0
        for j in range(n):
            ind = int(i-(n-1)/2+j)
            if ((ind>-1) and (ind<length)):
                 current += inp[ind]/n
        output[i] = current
    return output
```

```python
maf3 = movingAverageFilter(dataList,3)
maf5 = movingAverageFilter(dataList,5)
maf7 = movingAverageFilter(dataList,7)
```

```python
figure(figsize=(15, 7))

plt.subplot(2,1,1)
plt.plot(filteringData2)
plt.plot(maf3)
plt.axis('off')
plt.subplot(2,1,2)
plt.plot(filteringData3)
plt.plot(maf5)
plt.axis('off')

figure(figsize=(15, 7))

plt.subplot(2,1,1)
plt.plot(filteringData4)
plt.plot(maf7)
plt.axis('off')
```

```python
data1 = data.iloc[43]
data2 = data.iloc[24]
data3 = data.iloc[78]
```
